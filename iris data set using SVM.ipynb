```python
from sklearn.datasets import load_iris
iris = load_iris()
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#lets look at the basic features 
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
y = df['target'].values
X = df.drop(['target'], axis = 1)
```

```python
#lets see how many flowers are setosa, verginica, vericolor
df[df.target == 0].head()
```

```python
#LETS ADD ONE MORE COL AS FLOWERS_NAMES
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

# Explanation of the Above Cell

So what we did is added a new col as seen. SO for each value(i.e. x) in my target it will return 

the index  value in the target_names 


E.G

If the target value is 2 it will return the  index value from target_names i.e. virginica

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3 )
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_test[:10]
```

```python
model.predict(X_test[:10])
```

```python
y_predicted = model.predict(X_test)
```

# CONSUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test ,y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize =(10, 7))
sn.heatmap(cm, annot = True)#sq 
plt.xlabel("Predicted", fontsize = 20)
plt.ylabel("Truth", fontsize = 20)
plt.show()
```
